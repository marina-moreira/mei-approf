### **Introducing the motion activity recognition problem** <a class="anchor" id="motiv"></a>

#### Motivation

Motion activity recognition using sensory data is a crucial task for numerous applications, including healthcare monitoring, sports analytics, and human-computer interaction. The classification of activities such as sitting, jogging, walking, and standing through time series data collected from motion sensors poses a significant challenge. Accurate classification is vital for developing intelligent systems that can interact seamlessly with humans and monitor their activities in real-time.

This project aims to explore the use of Recurrent Neural Networks (RNNs), particularly LSTM and traditional RNN layers, to classify different motion activities from the MotionSense dataset. The primary goal is to achieve high accuracy in identifying motion activities by experimenting with various RNN architectures and time windows.

#### Objectives
- **Develop, implement, and test RNN models** to classify motion activities based on time series data from motion sensors;
- **Evaluate and compare the performance** of different RNN architectures, including LSTM and traditional RNN layers;
- **Analyze the impact of different time windows** on the model's classification performance;
- **Write a scientific paper** documenting the methodology, experiments, results, and conclusions.

#### Datasets
The dataset used for this project is from Kaggle, containing sensory data on various activities:
[MotionSense Dataset : Smartphone Sensor Data - HAR](https://www.kaggle.com/datasets/malekzadeh/motionsense-dataset?select=A_DeviceMotion_data)

#### Tools and Libraries
- **Programming Language:** Python
- **Deep Learning Framework:** TensorFlow\Keras
- **Notebook Environment:** Jupyter Notebook
Here you can get familiarized with the content more properly (the respective CRISP-DM stages of the project are indicated in parentheses):

<h2 style="background-color:#C2B4B9;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>Problem Definition </centre></strong></h2>

Motion activity classification is challenging due to the variety and similarity of activities. The dataset includes time series data of different activities, and the objective is to train an RNN model that can accurately classify these activities.

#### Specific Goals
+ Achieve an average accuracy of over 80% for each selected activity;
+ Evaluate the impact of different RNN layers such as LSTM and traditional RNNs on classification performance;
+ Analyze the effect of varying time windows on the model's ability to classify different motion activities;
+ Compare the performance of different combinations of RNN layers and time windows.

#### Questions to be Solved Through the Notebook
1. How does the choice of RNN layers, such as LSTM and traditional RNNs, affect the classification performance? 
2. What is the impact of varying time windows on the model's ability to classify different motion activities? 
3. How do different combinations of RNN layers and time windows influence the model's overall accuracy and robustness? 


<h2 style="background-color:#C2B4B9;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>Methodology </centre></strong></h2>

The project methodology includes the following steps:

**1.** **Data Understanding:**
   - Overview of the Dataset;
   - Initial Data Exploration;
   - Summary Statistics and Data Visualization; 
   - Missing Values and Data Quality.

**2.** **Data Preprocessing:**
   - Time series normalization;
   - Splitting data into training, validation, and test sets.

**3.** **Model Development:**
   - Designing different RNN architectures (e.g., LSTM, traditional RNN).
   - Experimenting with various time windows.

**4.** **Training and Evaluation:**
   - Training the models on the preprocessed data.
   - Evaluating model performance using metrics such as accuracy, precision, and recall.

**5.** **Analysis:**
   - Evaluating the impact of different RNN layers and time windows.
   - Analyzing overfitting and applying regularization techniques if necessary.

<h2 style="background-color:#C2B4B9;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>Expected Results </centre></strong></h2>

#### Expected Results
- **Model Accuracy:** Aim for an average accuracy of over 80% per selected activity.
- **Impact Analysis:** Detailed analysis of the impact of different RNN layers and time windows.
- **Comparison:** Insights into the robustness of different model architectures under various conditions.

#### Best and Worst Case Scenarios
- **Best-Case Examples:**  Time series segments where the model accurately classifies the motion activity.
- **Worst-Case Examples:** Time series segments where the model misclassifies the activity, including analysis of potential reasons.

<h2 style="background-color:#C2B4B9;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;"><strong><centre>Conclusion </centre></strong></h2>

* Data exploration results indicate the MotionSense dataset as high-quality and suitable for training RNN models for motion activity classification.

* Enhancing the network's architecture by experimenting with different RNN layers and time windows improves the model's accuracy and robustness in classifying motion activities.

In summary, there is potential for further improvement in model performance. Additional advancements can be achieved by optimizing model hyperparameters, employing stronger data augmentation techniques, and implementing advanced regularization methods.
