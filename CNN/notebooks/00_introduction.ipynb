### Introducing the bird species recognition problem <a class="anchor" id="motiv"></a>

#### Motivation

Bird species recognition is a complex task that challenges the visual capabilities of both human experts and computer algorithms. The classification of birds by species using imagery data collected from aerial surveys is an intriguing and significant problem. Bird populations serve as vital biodiversity indicators, making accurate data collection essential for ecologists. Reliable identification of bird species is crucial for monitoring biodiversity, understanding ecological dynamics, and implementing conservation strategies.

Furthermore, bird species recognition is particularly beneficial for companies involved in the development of wind farms that produce renewable energy. The construction of these wind farms requires thorough risk assessments to mitigate bird collisions, which pose a significant threat to many bird species worldwide. By accurately identifying bird species, we can better assess and minimize the risks associated with wind farm installations, thus contributing to the conservation of vulnerable bird populations while promoting sustainable energy development.

This project focuses on developing a Deep Learning model to classify various species of birds using images. The primary goal is to achieve high accuracy in identifying bird species using CNNs.

#### Objectives
- **Develop, implement, and test Deep Learning models** to solve practical classification problems.
- **Evaluate and compare the performance** of different CNN architectures.
- **Write a scientific paper** documenting the methodology, experiments, results, and conclusions.
