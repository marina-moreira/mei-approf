![Dataset Cover](https://storage.googleapis.com/kaggle-datasets-images/1209061/2019994/a8350fbe2c4118a2d426fedec0a61b04/dataset-cover.jpg?t=2021-03-13-18-29-09)

The Sports Image Classification dataset, available on Kaggle, is a comprehensive collection of high-quality images spanning 100 different sports categories. This dataset is meticulously curated to ensure it is free from duplicates and near-duplicates, providing a robust foundation for developing and testing image classification models. Each image is resized to 224x224 pixels and converted to jpg format, making it compatible with popular deep learning frameworks such as Keras.<br>
<br>

<h1 style="background-color:#C2B4B9;color:white;text-align: center;padding-top: 5px;padding-bottom: 5px;border-radius: 15px 50px;">
<strong><centre>Overview Dataset </centre></strong></h1>

The dataset is divided into training, validation, and test sets, ensuring a clear separation to prevent data leakage. It contains 13,493 training images, 500 validation images, and 500 test images. Each image is carefully processed and resized to 224x224x3 format. A CSV file is included for those who prefer to create their own train, test, and validation datasets using `ImageDataGenerator.flow_from_dataframe`.
